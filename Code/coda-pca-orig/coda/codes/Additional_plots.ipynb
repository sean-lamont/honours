```python
import numpy as np
import matplotlib.pyplot as plt
import os 
from pathlib import Path
```

```python
p = Path(os.getcwd()).parents[3]
PATH = os.path.join(os.path.join(p, "thesis"), "figs")
```

```python
#PCA example
mean = [0, 0]
cov = [[100, 0], [0, 10]]  # diagonal covariance

x, y= np.random.multivariate_normal(mean, cov, 15).T

fig = plt.figure(figsize = (10, 5))
plt.stem(x, y, 'y')#
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.title("Points from a 2D Gaussian ($\sigma_x^2 = 100, \sigma_y^2 = 10, cov(x,y)=0$), with 1D projection ")
plt.savefig(os.path.join(PATH, "pca_gaussian.svg"))
```
