```python
import pandas as pd
import numpy as np
```

```python
#read in cattle metadata to use as class labels/regression targets
meta = pd.read_excel('RUG Meta.xlsx')
#remove useless columns
meta = meta[meta.columns[0:4]]
#remove unknown entry
meta = meta[0:-1]
```

```python
#transpose and use rename to get the sample ID as the row labels, allowing a join with features 
meta=meta.T
meta = meta.rename(columns=meta.iloc[1])
```

```python
#transpose again to return to (sample, features) format
meta = meta.T
```

```python
#Remove sample (as it is now row labels) and Sex since they are all male
meta = meta.drop('Sample',axis=1)
meta = meta.drop('Sex',axis=1)
```

```python
#read in abundance data
abundances = pd.read_excel('RUG data.xlsx')
```

```python
#select only the read depths per cattle and the associated identifier
abundances = pd.concat([abundances[abundances.columns[0]] , abundances[abundances.columns[4:286]]], axis=1)
```

```python
#transpose to get standard (sample, features) shape
abundances = abundances.T
```

```python
#set sample ID as row label and remove it as a data column
abundances.rename(columns=abundances.iloc[0])
abundances = abundances[1:]
```

```python
#join meta and abundances based on sample ID
full = abundances.join(meta)
full
```

```python
#store as hdf5 for quick serialisation and loading
full.to_hdf('rug.h5', 'rug')
```
